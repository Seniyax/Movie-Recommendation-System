```python
ratings_df = pd.read_csv('/content/CF_data.csv')
ratings_df.head()
```

```python
ratings_df["Rating"].unique()
```

```python
from surprise import Reader,Dataset

reader = Reader(rating_scale=(1,5))

data = Dataset.load_from_df(ratings_df,reader)
```

```python
from surprise.model_selection import train_test_split

trainset,testset = train_test_split(data,test_size=0.2,random_state=42)
```

```python
from surprise import SVD
from surprise import accuracy

model = SVD()
model.fit(trainset)
```

```python
prediction = model.test(testset)
```

```python
error = accuracy.rmse(prediction)
```

```python
cf_scores = []

user_id = 1
ratings_df["UserID"] == user_id

movie_ids = ratings_df['MovieID']

for movie_id in movie_ids:
  predict = model.predict(user_id,movie_id)
  cf_scores.append((movie_id,predict.est))


cf_scores = pd.DataFrame(cf_scores,columns=["MovieId","CF_scores"])
cf_scores.head()
```

```python
cf_scores.to_csv("cf_scores_user_1.csv",index=False)
```
